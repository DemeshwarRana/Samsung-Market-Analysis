```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import pandas as pd
```

```python
data={
    'Days':['Sun','Mon','Tue','Wed','Thu','Fri','Sat']*2,
    'XP':[200,50,70,260,500,450,350,20,60,80,500,380,310,290],
    'User':['Demeshwar']*7+['William']*7
}
df=pd.DataFrame(data)
sns.set_theme(style='dark',rc={
    'axes.facecolor':'black',
    'figure.facecolor':'black',
    'grid.color':'grey'
})
plt.figure(figsize=(10,6))
ax=sns.lineplot(data=df,x='Days',y='XP',hue='User',marker='o',markersize='12',palette=['grey','skyblue'],linewidth=2.5,zorder=7)
total_demeshwar=df[df['User']=='Demeshwar']['XP'].sum()
total_william=df[df['User']=='William']['XP'].sum()
plt.text(6,650,f'{total_demeshwar}XP',color='white',fontsize=11,fontweight='bold',family="cursive")
plt.text(6,620,f'{total_william}XP',color="skyblue",fontweight="bold",fontsize=14,family='cursive')
plt.text(0,310,'(Target XP)',color='white')
plt.title('XP this week',fontweight='bold',fontsize=16,color='white',loc='left')
plt.xlabel('Days',color='white',fontsize=12,fontweight='bold',family='cursive')
plt.ylabel('XP',color='white',fontsize=12,fontweight='bold',family="cursive")
plt.tick_params(axis="both",labelcolor='white',labelsize=12)
plt.axhline(y=300,color='red',linestyle='--',linewidth=3,alpha=1,zorder=5)
sns.despine(left=True,bottom=True)
plt.ylim(0,600)
plt.legend(frameon=False,loc='upper left',labelcolor='white',handlelength=0,prop={
    'family':'cursive',
    'size':12,
    'weight':'bold'
})
plt.grid(axis='y',color='grey',linewidth=2,alpha=0.8,zorder=0)
plt.savefig('duolingograph.png')
plt.show()
```
