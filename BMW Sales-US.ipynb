```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import pandas as pd
```

```python
quarters = ['Q1 2022', 'Q2 2022', 'Q3 2022', 'Q4 2022', 'Q1 2023', 'Q2 2023', 'Q3 2023']
bev_sales = [500, 1500, 4500, 9000, 6800, 11800, 13500]
bev_share = [0.5, 1.5, 5.0, 9.0, 8.2, 13.5, 15.8] 
fig = make_subplots(specs=[[{"secondary_y": True}]])
fig.add_trace(go.Bar(x=quarters,
        y=bev_sales,name="BMW BEV Sales",marker_color='#222222',marker_line=dict(width=1, color='white')
    ),
    secondary_y=False
)
fig.add_trace(go.Scatter(x=quarters,y=bev_share,name="BEV share in BMW sales",mode='lines',
        line=dict(color='#d4fb00', width=4, shape='spline')
    ),
    secondary_y=True)
fig.update_layout(
    title_text="BMW BEV Sales - US",
    title_x=0.5,
    paper_bgcolor='#444444',
    plot_bgcolor='#444444',
    font=dict(color="white"),
    legend=dict(orientation="h",yanchor="bottom",y=-0.2,xanchor="center",x=0.5),
    margin=dict(l=50, r=50, t=80, b=50)
)
fig.update_yaxes(title_text="Units (k)",secondary_y=False, gridcolor='#666666',range=[0, 20000])
fig.update_yaxes(title_text="Share (%)", secondary_y=True, showgrid=False,range=[0, 20])
fig.show()
```
